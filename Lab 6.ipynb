```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([1.2,1.5,1.9,2.2,2.4,2.5,2.8,3.1, 3.3, 3.7, 4.2, 4.4])
y = np.array([1.7, 2.4, 2.3, 3.1, 3.7, 4.2, 4.4, 6.1, 5.4, 5.7, 6.4, 6.2])

c = 1.1
m1, m2, m3 = 0.1, 1.5, 0.8

y1 = m1*x + c
y2= m2*x + c
y3 = m3*x + c

def mse(ypred, y):
    return np.sum(abs(y-ypred))

e1 = mse(y1,y)
e2 = mse(y2, y)
e3 = mse(y3, y)
e = np.array([e1,e3, e2])
b = np.array([0.1,0.8, 1.5])

plt.plot(b,e, color="pink")
plt.show()

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.scatter(x, y)
plt.show()
```

# Q2

```python
import numpy as np

X = np.array([1,2,3,4,5,6,7,8])
Y = np.array([0,0,0,0,1,1,1,1])

l = 0.01
iters = 60
losses = []
w = 0  

for _ in range(iters):
    yhat = 1 / (1 + np.exp(-X * w))
    grad = np.dot(X, (yhat - Y)) / len(Y)
    w -= l * grad
    loss = -np.mean(Y * np.log(yhat) + (1 - Y) * np.log(1 - yhat))
    losses.append(loss)

plt.plot(losses)
plt.show()
```

# Q3

```python
X = np.array([1, 2, 4, 3, 5]) 
Y = np.array([1, 3, 3, 2, 5])  # Positive slope 

m = 0
c = 0

L = 0.01
epochs = 50
n = float(len(X))

errors = []

for i in range(epochs):
    Y_pred = m * X + c
    D_m = (-2 / n) * np.sum(X * (Y - Y_pred))
    D_c = (-2 / n) * np.sum(Y - Y_pred)
    m -= L * D_m
    c -= L * D_c
    error = np.mean((Y - Y_pred) ** 2)
    errors.append(error)

plt.scatter(X, Y)
ypred = m * X + c
plt.plot(X, Y_pred)
plt.show()

plt.plot(errors)
plt.show()
```
